```python
import cv2
import numpy as np
import random
import os
import csv

def generate_image_with_dots(image_size=(256, 256), max_dots=100):
    # Create a white image
    image = np.ones((image_size[0], image_size[1], 3), dtype=np.uint8) * 255
    
    colors = {
        'red': (0, 0, 255),
        'green': (0, 255, 0),
        'blue': (255, 0, 0)
    }
    
    counts = {
        'red': 0,
        'green': 0,
        'blue': 0
    }
    
    occupied_positions = set()

    def draw_dot(color, count_key):
        while True:
            dot_radius = random.randint(5, 10)
            x = random.randint(dot_radius, image_size[0] - dot_radius - 1)
            y = random.randint(dot_radius, image_size[1] - dot_radius - 1)
            if not any((x + dx, y + dy) in occupied_positions for dx in range(-dot_radius, dot_radius + 1) for dy in range(-dot_radius, dot_radius + 1)):
                cv2.circle(image, (y, x), dot_radius, color, -1)
                for dx in range(-dot_radius, dot_radius + 1):
                    for dy in range(-dot_radius, dot_radius + 1):
                        occupied_positions.add((x + dx, y + dy))
                counts[count_key] += 1
                break

    total_dots = random.randint(1, max_dots)
    colors_list = ['red', 'green', 'blue']
    for _ in range(total_dots):
        color_key = random.choice(colors_list)
        draw_dot(colors[color_key], color_key)
    
    # Normalize the counts
    total_count = counts['red'] + counts['green'] + counts['blue']
    if total_count > 0:
        counts['red'] = counts['red'] / total_count
        counts['green'] = counts['green'] / total_count
        counts['blue'] = counts['blue'] / total_count
    
    return image, counts

if __name__ == '__main__':
    if not os.path.exists('images'):
        os.makedirs('images')
    
    while True:
        try:
            number_of_images = int(input("Enter Number of Clock Images to generate: "))
            break
        except ValueError:
            print("Please enter a valid integer.")
    
    with open('labels.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['File Name', 'Red Count', 'Green Count', 'Blue Count'])
        for i in range(1, number_of_images + 1):
            image_with_dots, counts = generate_image_with_dots()
            image_path = f'images/dots_{i}.png'
            cv2.imwrite(image_path, image_with_dots)
            print(f"Generated {image_path} with counts: {counts}")
            writer.writerow([image_path, counts['red'], counts['green'], counts['blue']])
    
    print(f"Successfully generated {number_of_images} images")
```

```python
import os
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input, BatchNormalization

# Clear any previous GPU memory allocations
tf.keras.backend.clear_session()

# Set memory growth for GPU
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Load labels
labels_df = pd.read_csv('labels.csv')

# Parameters
IMG_SIZE = (256, 256)
BATCH_SIZE = 8
VALIDATION_SPLIT = 0.2

# ImageDataGenerator for data augmentation and preprocessing
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=VALIDATION_SPLIT,
    
)

# Creating training and validation generators
train_generator = datagen.flow_from_dataframe(
    labels_df,
    x_col='File Name',
    y_col=['Red Count', 'Green Count', 'Blue Count'],
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='raw',
    subset='training'
)

validation_generator = datagen.flow_from_dataframe(
    labels_df,
    x_col='File Name',
    y_col=['Red Count', 'Green Count', 'Blue Count'],
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='raw',
    subset='validation'
)

def build_regression_cnn(input_shape):
    inputs = Input(shape=input_shape)
    
    # Convolutional layers
    x = Conv2D(16, (3, 3), activation='relu')(inputs)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = Conv2D(32, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)
    
    # Fully connected layers
    x = Dense(32, activation='relu')(x)
    x = Dropout(0.5)(x)
   
    
    # Output layer
    outputs = Dense(3, activation='linear')(x)  # Predict 3 values for red, green, and blue dot counts
    
    return Model(inputs, outputs)

# Define a custom accuracy metric
def custom_accuracy(y_true, y_pred):
    tolerance = 1.0  # Define tolerance range
    abs_diff = tf.abs(y_true - y_pred)
    accurate_predictions = tf.less_equal(abs_diff, tolerance)
    return tf.reduce_mean(tf.cast(accurate_predictions, tf.float32))

# Build the model
model = build_regression_cnn(IMG_SIZE + (3,))
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.Huber(),
    metrics=['mae', custom_accuracy]
)

model.summary()

# Train the model
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=validation_generator
)
model.evaluate(validation_generator,batch_size=8)
# Save the model
model.save('dots_predictor.h5')

# Plot training & validation loss and accuracy values
plt.figure(figsize=(12, 5))

# Plot training and validation loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot training and validation MAE
plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Mean Absolute Error')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()

# Plot training and validation custom accuracy
plt.figure(figsize=(12, 5))
plt.plot(history.history['custom_accuracy'], label='Training Custom Accuracy')
plt.plot(history.history['val_custom_accuracy'], label='Validation Custom Accuracy')
plt.title('Custom Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Load the trained model
model = tf.keras.models.load_model('dots_predictor.h5', custom_objects={'custom_accuracy': custom_accuracy})

# Load test labels
test_labels_df = pd.read_csv('test_labels.csv')

# Parameters
IMG_SIZE = (256, 256)
BATCH_SIZE = 8

# ImageDataGenerator for data preprocessing (no data augmentation for test data)
test_datagen = ImageDataGenerator(rescale=1./255)

# Creating test generator
test_generator = test_datagen.flow_from_dataframe(
    test_labels_df,
    x_col='File Name',
    y_col=['Red Count', 'Green Count', 'Blue Count'],
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='raw',
    shuffle=False
)

# Make predictions on test data
predictions = model.predict(test_generator)

# Evaluate the model
mae = tf.keras.metrics.mean_absolute_error(test_generator.labels, predictions).numpy()
print(f'Mean Absolute Error: {mae}')

# Custom accuracy
tolerance = 0.1  # You can adjust the tolerance as needed
absolute_differences = np.abs(test_generator.labels - predictions)
accurate_predictions = np.all(absolute_differences <= tolerance, axis=1)
custom_accuracy = np.mean(accurate_predictions)
print(f'Custom Accuracy: {custom_accuracy}')

# Plot actual vs predicted values
plt.figure(figsize=(10, 5))
for i, color in enumerate(['Red Count', 'Green Count', 'Blue Count']):
    plt.scatter(test_generator.labels[:, i], predictions[:, i], label=f'{color} (R^2 = {np.corrcoef(test_generator.labels[:, i], predictions[:, i])[0, 1]:.2f})')
    plt.plot([0, 1], [0, 1], 'r--')  # Line of perfect prediction
plt.xlabel('Actual Count')
plt.ylabel('Predicted Count')
plt.legend()
plt.title('Actual vs Predicted Counts')
plt.show()
```
